```python
import pandas as pd
import os
```

```python
df_main_exp = pd.DataFrame()
df_main_stats = pd.DataFrame()
df_main_tuning = pd.DataFrame()
```

```python

for 
curr_dir = ('.')

for file in os.listdir(curr_dir):

    if not file.endswith('.csv'):
        continue
    df = pd.read_csv(file)
    if file.split('_')[-1].startswith('experience'):
        
        df_main_exp = pd.concat([df_main_exp, df], axis=0)
    elif file.split('_')[-1].startswith('statistics'):
        df_main_stats = pd.concat([df_main_stats, df], axis=0)
    
    elif file.split('_')[-1].startswith('tuning'):
        df_main_tuning = pd.concat([df_main_tuning, df], axis=0)
```

```python
df_main_exp
```

```python

df = df_main_tuning
all_zeros_rows = df[(df.iloc[:, 2:12]  == 0).all(axis=1)]  
non_zero_rows = df[(df.iloc[:, 2:12] != 0).any(axis=1)]  
```

```python
all_zeros_rows
```

```python
from scipy.stats import mannwhitneyu
min_length = min(len(all_zeros_rows), len(non_zero_rows))

mannwhitneyu(all_zeros_rows.iloc[:min_length]['change'], non_zero_rows.iloc[:min_length]['change'])
```
